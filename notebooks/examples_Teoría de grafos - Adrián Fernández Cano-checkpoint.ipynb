##Definición de grafo
Un grafo <b>G = (V, E)</b> es una par ordenado donde <b>V</b> es un conjunto normalmente finito de vértices(nodos) unidos
por enlaces (aristas). <b>E</b> es el conjunto de pares (aristas) que relacionan a estos vértices.

- Si <b>E</b> es un conjunto de <b>pares no ordenados</b> de elementos de V es un <b>grafo no dirigido</b>.
- Si <b>E</b> es un conjunto de <b>pares ordenados</b> de elementos de V es un <b>grafo dirigido</b>.
    - Dada una arista (a, b), a es su nodo inicial y b su nodo final.

##Definiciones
- Se llama <b>orden</b> del grafo G a su número de vértices, |V|.
- El <b>grado de un vértice</b> es el número de aristas que lo tienen como extremo.
- El <b>grado de un grafo</b> es igual a la suma de los grados de todos sus vértices. Es el doble del número de aristas o arcos.
    - Propiedad:
        El grado de un grafo va a ser igual al doble de sus aristas 
            - grado = 2|E|
            
- Un <b>bucle</b> es una arista que relaciona al mismo nodos. El nodo inicial y final coinciden.
- Dos o más aristas que relacionan al mismo par de vertices se denominan <b>aristas múltiples o pararelas</b>.
- Decimos que dos aristas son adyacentes si tienen un extremo en común.
- Un vertice u es <b>adyacente</b> a otro vertice w si el grafo contiene una arista (u,w) que los une.
- Un <b>nudo aislado</b> es un vertice (nodo) que no está unido a ningún otro vértice o nodo.

```python
"""Crear multigrafo vacío (sin nodos ni aristas)"""
grafo1 = nx.MultiGraph() 

"""Crear un 'bundle' de nodos y añadirlos"""
bundle = nx.path_graph(5)
grafo1.add_nodes_from(bundle)

""" Establecer las aristas"""
grafo1.add_edges_from([(0,1), (1,2), (2,3), (3,1), (0,3), (0,3), (1,1)])
draw(grafo1, show='ipynb')
```

- El <b>orden</b> del grafo anterior será -> numVertices = grafo.number_of_nodes() -> 4
- El <b>grado</b> del grafo anterior será -> numAristas = grafo.number_of_edges() -> 7
- En el nodo 1 tenemos un <b>bucle</b>.
- Entre el nodo 0 y 3 tenemos <b>aristas múltiples</b>.
- El vertice 2 es <b>adyacente</b> al vertice 3.
- La arista (2, 3) es adyacente a la arista (1,2).
- El vértice 4 es un <b>nudo aislado</b>.

##Grafo dirigido o digrafo
Es un tipo de grafo en el cual las aristas tienen un <b>sentido definido</b>.
G = (V, E)
A los elementos del conjunto E se les denomina <b>arcos</b>.
- Un arco e = (x,y) se considera dirigido desde x hacía y.
    - y se denomina origen o cabeza.
    - x se denomina extremo o cola.
- Al arco e2 = (y,x) se le denomina arco invertido de e.
- Dos arcos son <b>adyacentes</b> si el extremo de uno de ellos es el origen del siguiente.

En caso de ser un digrafo el grado de un vértice:
   - <b>Grado de entrada</b>, al número de arcos que llegan al vértice.
   - <b>Grado de salida</b>, al número de arcos que parten del vértice.

Pero el grado del vértice se seguirá calculando como antes y en este caso será:
   - grado vertice = grado entrada + grado salida

En el siguiente ejemplo el arco (0,3) y el arco (3,2) son adyacentes.

```python
digrafo = nx.DiGraph()
digrafo.add_edges_from([(0,3), (3,2), (0,1), (2,1), (1,2), (3,1)])
draw(digrafo, show='ipynb')
```

##Digrafo completo

distintos existe un arco.

El siguiente ejemplo <b>NO</b> es un digrafo completo debido a que no existe el arco (0,4) o el arco (4,0), además
de que existen dos arcos entre los vértices 1 y 3.

```python
digrafoNoCompleto = nx.DiGraph()
digrafoNoCompleto.add_edges_from([(0,1), (1,3), (2,3), (1,2), (3,4), (2,4),(3,1)])
draw(digrafoNoCompleto, show='ipynb')
```

##Representación de grafos
Las dos representaciones principales son la <b>matriz de adyacencia</b> y la <b>lista de adyacencia</b>.
- <b>Matriz de adyacencia (MA)</b>: Sea un grafo G = (V, E), con |V| = n. La matriz de adyacencia de G es de tamaño n*n, las
filas y las columnas hacen referencia a los vértices. La celda MA[i, j] almacena el número de aristas entre el vertice i y j. Si 
su valor es 0 significa que no existe arista entre esos vertices.

El ejemplo siguiente muestra una matriz parecida a la mostrada en el multigrafo pero creada utilizando una matriz de adyacencia.
El valor de la celda MA[0,3] vale 3 para mostrar tres aristas entre los vertices 0 y 3:

```python
matrizAdyacencia = np.matrix([[0,1,0,3],
                             [1,1,1,1],
                             [0,1,0,1],
                             [3,1,1,1]])

"""Para poder mostrar las aristas multiples se debe poner 'parallel_edges = True' y  'create_using = nx.MultiGraph()'"""
grafoMatrizAdyacencia = nx.from_numpy_matrix(matrizAdyacencia, parallel_edges = True, create_using = nx.MultiGraph())
draw(grafoMatrizAdyacencia, show='ipynb')
```

##Camino
Se denomina <b>camino, trayectoria o recorrido</b> a una secuencia de vértices dentro de un grafo tal que exista una arista entre cada par de vértices.
Dos caminos son <b>independientes</b> si no tienen ningún vértice en común excepto el primero y el último.
- Un camino es <b>simple</b> si todas sus aristas son distintas (no repite vértices en su recorrido).
- La <b>longitud</b> de un camino es el número de aristas que usa dicho camino, contando las aristas que se repitan varias veces.
- La <b>distancia</b> de un camino (o entre dos vertices de un grafo) es el número de vértices mínimo que debe recorrerse para unirlos y es n/2.

###Circuito
Un <b>circuito</b> es un camino cerrado que no repite aristas.

El siguiente grafo de n vértices es un ejemplo de camino de orden n.
- Orden: n = 4
- Aristas (longitud): n-1 = 3
- Distancia: n/2 = 4/2 = 2



```python
grafo = nx.Graph() 
bundle = nx.path_graph(4)
grafo.add_nodes_from(bundle)
grafo.add_edges_from([(0,1), (1,2), (2,3)])
draw(grafo, show='ipynb')
```

##Potencias de la matriz de adyacencia
La matriz de adyacencia de un grafo elevada a P nos determina el número de caminos de longitud P distintos.
Si en la matriz resultante nos encontramos un 0 significaría que entre esos dos vértices no existe ningún
camino de P tramos(longitud)

Ejemplo, usaremos la matriz de adyacencia usada anteriormente.

```python
draw(grafoMatrizAdyacencia, show='ipynb')
```

A continuación elevamos a 2 (P = 2) la matriz de adyacencia para obtener el número de caminos de longitud 2 distintos entre
cada par de nodos.

```python
matriz = np.linalg.matrix_power(matrizAdyacencia, 2)
print(matriz)
```

Interpretando el resultado, significa por ejemplo que el número de caminos de longitud 2 distintos desde el nodo 1 al nodo 3
es igual a matriz[1, 3] = 6

##Grafo conexo
Un grafo es <b>conexo</b> si para cada par de vértices existe un camino que los une, es decir, si para cualquier par de
vértices (a,b)existe un camino posible desde a hacia b.

```python
grafoConexo = nx.Graph() 
bundle = nx.path_graph(5)
grafoConexo.add_nodes_from(bundle)
grafoConexo.add_edges_from([(0,1), (0,2), (2,4), (2,3),(1,3)])
draw(grafoConexo, show='ipynb')
```

##Grafo ponderado, etiquetado o pesado
Un grafo es <b>ponderado</b> cuando sus aristas contienen datos (etiquetas). Una etiqueta puede ser un nombre, un coste o un valor
de cualquier tipo de dato.
El <b>peso</b> total de un grafo ponderado es la suma de los pesos de sus aristas.

```python
grafoPonderado=nx.Graph()
grafoPonderado.add_node(1,pos=(1,1))
grafoPonderado.add_node(2,pos=(2,2))
grafoPonderado.add_node(3,pos=(1,0))
grafoPonderado.add_edge(1,2,weight=10)
grafoPonderado.add_edge(1,3,weight=9)
grafoPonderado.add_edge(2,3,weight=20.0)
pos=nx.get_node_attributes(grafoPonderado,'pos')
nx.draw_networkx(grafoPonderado,pos)
labels = nx.get_edge_attributes(grafoPonderado,'weight')
plt.axis('off')
nx.draw_networkx_edge_labels(grafoPonderado,pos,edge_labels=labels)
```

##Isomorfismo de grafos
Dos grafos G1 y G2 son isomorfos si existe una funcíon entre los vértices de ambos conjuntos que cumpla:
f:V1 -> V2
   - Que sea <b>biyectiva</b>.
   - Las imágenes de dos vértices adyacentes son adyacentes. 
Si se cumple se dice que f es un isomorfismo de grafos entre G1 y G2.

Veamos un ejemplo, utilizo el grafo generado anteriormente con una matriz de adyacencia.

```python
draw(grafoMatrizAdyacencia, show='ipynb')
```

Creo una nueva matriz de adyacencia 

```python
matrizAdyacencia2 = np.matrix([[0,1,0,1],
                               [3,1,1,1],
                               [0,1,0,3],
                               [1,1,1,1]])



"""Para poder mostrar las aristas multiples se debe poner 'parallel_edges = True' y  'create_using = nx.MultiGraph()'"""
grafoMatrizAdyacencia2 = nx.from_numpy_matrix(matrizAdyacencia2, parallel_edges = True, create_using = nx.MultiGraph())
draw(grafoMatrizAdyacencia2, show='ipynb')
```

Comprobamos is entre ambos grafos existe un isomorfismo:

```python
nx.is_isomorphic(grafoMatrizAdyacencia, grafoMatrizAdyacencia2)
```

##Grafo bipartido
Un grafo G = (V,E), no dirigido, será <b>bipartido</b> si sus vértices se pueden separar en dos conjuntos disjuntos A y B tal
que se cumpla:
    - A U B = V
    - A ∩ B = ø
De manera que las aristas solo pueden conectar vértices de un conjunto con vñertices del otro, suelen representarse
gráficamente en dos columnas de vértices.
Si coloreamos los vértices de ambos conjuntos A y B de rojo y verde por ejemplo, cada arista tendrá un vértice
rojo y otro verde.

   - Un grafo G es <b>bipartido</b> si, y sólo si, G no tiene ciclos de longitud impar.
    

```python
grafoBipartido = nx.Graph() 
bundle = nx.path_graph(6)
grafoBipartido.add_nodes_from(bundle)
grafoBipartido.add_edges_from([(0,3), (1,3), (1,5), (2,3),(2,5),(3,4)])
draw(grafoBipartido, show='ipynb')
```

En el ejemplo anterior se podrían ver los nodos 0,1,2,4 de color rojo y los nodos 3 y 5 de color verde.
Verificamos que es bipartido:

```python
nx.is_bipartite(grafoBipartido)
```

##Grafo bipartido completo
Será grafo <b>bipartido completo</b> aquel grafo que además de ser bipartido tiene todos los vértices de la 
partición V1 conectados a todos los vértices de la partición V2, es decir, cada vértice se conecta con todos
los vértices del otro conjunto.

```python
grafoBipartidoCompleto = nx.Graph() 
bundle = nx.path_graph(6)
grafoBipartidoCompleto.add_nodes_from(bundle)
grafoBipartidoCompleto.add_edges_from([(0,3), (0,5), (1,3), (1,5), (2,3),(2,5),(3,4),(4,5)])
draw(grafoBipartidoCompleto, show='ipynb')
```

##Grafo plano
Un grafo o multigrafo será un <b>grafo plano</b> si se puede dibujar en el plano sin que se crucen sus arcos se llama plano.
Los grafos K5 y K3,3 son los grafos no planos que nos permiten caracterizar el resto de los grafos no planos.
   - <b>Teorema de Whitney</b>
       Un grafo es plano si y solo si no contiene a ningún subgrafo contraible a K5 o K3,3.
       Una contracción es el resultado de eliminar una arista uniendo sus vértices en un único vértice.

```python
grafoBipartidoCompleto = nx.Graph() 
bundle = nx.path_graph(3)
grafoBipartidoCompleto.add_nodes_from(bundle)
grafoBipartidoCompleto.add_edges_from([(0,1), (1,2), (2,3), (3,0),(0,2), (1,3)])
draw(grafoBipartidoCompleto, show='ipynb')
```

##Grafo euleriano
Llamamos <b>camino o circuito euleriano</b> a un camino o circuito que contiene todas las aristas, apereciendo cada una
de ellas exactamente una vez.

Entonces llamamos <b>grafo euleriano</b> a un grafo que admite un circuito euleriano.

En el ejemplo siguiente el camino euleriano sería:
    (5,2),(2,0),(0,3),(3,2),(2,1),(1,0),(0,4),(4,1),(1,3),(3,4),(4,5)

```python
grafoEuleriano = nx.Graph() 
bundle = nx.path_graph(5)
grafoEuleriano.add_nodes_from(bundle)
grafoEuleriano.add_edges_from([(0,1), (0,2), (0,3), (1,2), (1,3),(1,4), (4,5), (5,2),(4,3),(2,3)])
draw(grafoEuleriano, show='ipynb')
```

##Camino hamiltoniano
Un <b>camino hamiltoniano</b> es un camino de un grafo que visita todos los vértices del grafo una sola vez.
Si además el último vértice visitado es adyacente al primero, el camino es un ciclo hamiltoniano.
