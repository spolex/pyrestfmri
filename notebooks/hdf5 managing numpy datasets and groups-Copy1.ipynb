```python
import h5py
import numpy as np
import matplotlib.pyplot as plt
```

```python
h5_ts_file = h5py.File('elekin_phd_et.hdf5','r')
```

```python
h5_ts_file.keys()
```

```python
group = h5_ts_file.get('cerebellum_ts')
group.keys()
```

```python
[item for item in group]
```

```python
timeseries = []
for item in group:
    print(item)
    timeseries.append(np.array(group.get(item)))
timeseries = np.array(timeseries)
```

```python
for i,title in enumerate(group):
    fig = plt.figure()
    plt.title(title)
    plt.plot(timeseries[i])
    plt.xlabel('')
    plt.ylabel('')
    plt.show()
    plt.close()
```

```python
h5_ts_file.close()
```
