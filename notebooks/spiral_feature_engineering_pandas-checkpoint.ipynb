```python
features_labeled = features.join(origin[["subject_id","labels"]].drop_duplicates().set_index('subject_id'))
```

```python
fts = [features_labeled.columns[0:10]]
np.append(fts, 'labels')
```

```python
sns.pairplot(data=features_labeled[np.append(fts, 'labels')],hue="labels")
```

```python
# compute correlation matrix
corr = features_labeled.corr()

# Generate a mask for the upper triangle
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.9, center=0,
            square=True, linewidths=.80, cbar_kws={"shrink": .5})
```
