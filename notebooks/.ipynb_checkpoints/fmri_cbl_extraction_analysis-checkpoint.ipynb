```python
import pandas as pd
%matplotlib inline
from matplotlib import pyplot as plt
import seaborn as sns
sns.set()
```

```python
h5_file = '/results/00-EXTRACCION-CEREBELO/elekin_fmri_ts_pandas.hdf5'
et_key = 'cbl/et'
ct_key = 'cbl/ct'
```

```python
ct = pd.read_hdf(h5_file, ct_key)
ct_subjects = pd.unique(ct.sort_values('id')['id'])

et = pd.read_hdf(h5_file, et_key)
et_subjects = pd.unique(et.sort_values('id')['id'])

subjects = et_subjects
df = et

indx = 2
```

```python
p = df[df["id"] == et_subjects[indx]].plot(figsize=(15,15), title="Cerebellum extract for subject "+et_subjects[indx])
```

```python
for subj in subjects:
    p = df[df["id"] == subj].plot(figsize=(15,15), title="Cerebellum extract for subject "+subj)
```

```python
fig, axes = plt.subplots(nrows=7, ncols=4, figsize=(15,15))
df = df.loc[:,df.columns != 'label']
p = df[df["id"] == subjects[indx]].plot(subplots=True, ax=axes, fig=fig, title="Cerebellum extract for subject "+subjects[indx])
```
