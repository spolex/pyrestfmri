```python
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
#sci py libraries
import numpy as np
import matplotlib.pyplot as plt
import nibabel as nb
import os
import h5py
```

Load data previously preprocessed

```python
x_size = 61
y_size = 61
n_slice = 73
n_volumes = 158
#158, 61, 73, 61

data_path = '/home/elekin/datos-niix/preproc/'
files = os.listdir(data_path)
node = 'bandpass_filter'
file_name = 'detrend_regfilt_filt.nii.gz'

fmri_files_path = list(map(lambda x: os.path.join(data_path,x,node,file_name),files))
X_train = np.array(list(map(lambda x: nb.load(x).get_data(),fmri_files_path)))
print(np.max(X_train),np.min(X_train))
X_train.shape
```

```python
os.listdir(data_path)
```

```python
from keras.models import Sequential
from keras.layers import Dense,Dropout
from keras.utils import to_categorical
```

Requisito Keras: tipo de las clases categorico

```python
Y_train = to_categorical([0,1,0,1])
#test_labels = to_categorical(test_labels)
```

Aplanar y normalizar los datos para utilizarlos de entrada de la NN

```python
from sklearn import preprocessing
X_train = preprocessing.scale(X_train.T.reshape((X_train.shape[0], -1)))
#X_test = X_test.reshape((X_test.shape[0], -1))
```

```python
for i in range(X_train.shape[0]):
    plt.plot(X_train[i,5000:5157])
```

En caso de querer normalizarlo entre 0 y 1

```python
#https://scikit-learn.org/stable/modules/preprocessing.html
min_max_scaler = preprocessing.MinMaxScaler()
X_train_minmax = min_max_scaler.fit_transform(X_train)
for i in range(X_train_minmax.shape[0]):
    plt.plot(X_train_minmax[i,5000:5157])
```

```python
X_train_minmax.shape
```

```python
model = Sequential()
model.add(Dense(8, input_shape=(X_train_minmax.shape[-1],), activation='relu'))
model.add(Dense(16, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dropout(0.15))
model.add(Dense(2, activation='softmax'))

model.compile(loss='categorical_crossentropy',
              optimizer='rmsprop',
              metrics=['accuracy'])
```

```python
history = model.fit(X_train_minmax, Y_train, epochs=5, batch_size=16, validation_split=0.2)


# Plot training & validation accuracy values
plt.plot(history.history['acc'])
plt.plot(history.history['val_acc'])
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')
plt.show()

# Plot training & validation loss values
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')
plt.show()
```
