Localización de los ficheros de parámetros

```python
data_path = '/home/elekin/datos-niix/preproc/'
files = list(filter(lambda dir: '_session_id_1_subject_id' in dir,os.listdir(data_path)))
node = 'mcflirt'
#file_name = 'sub-C024_ses-1_task-rest_bold_trim_st_mcf.nii.gz.par'
file_name = '_ses-1_task-rest_bold_trim_st_mcf.nii.gz.par'


par_files = np.array(list(map(lambda file: os.path.join(data_path,file,node,'sub-'+file[-4:]+file_name),files)))
```

```python
par_list = np.empty([102,158,6])
par_list.shape
```

```python
import h5py
h5_ts_file = h5py.File('elekin_phd_et.hdf5','r')
h5_ts_file.keys()
```

```python
par_group=h5_ts_file['motion_pars']
par_group.keys()
```

```python
for i,item in enumerate(par_files):
    new = np.loadtxt(item)
    if new.shape[0] == 158:
        par_list[i]=new
        par_group.create_dataset(files[i],data=new,compression="gzip", compression_opts=9)
```

```python
par_list[:,:,:2].shape
```

```python
for i,title in enumerate(files):
    fig = plt.figure()
    plt.title(title)
    plt.subplot(2,2,1)
    plt.title(title[-4:]+': traslación')
    plt.plot(par_list[i,:,:3])
    plt.subplot(2,2,2)
    plt.title(title[-4:]+': rotacion')
    plt.plot(par_list[i,:,4:])
    plt.xlabel('')
    plt.ylabel('')
    plt.show()
    plt.close()
    print(par_list[i].shape)
```
