## Numerical features
Numerical variables are simply those for which the values are numbers. The first thing that we do when we have numerical variables is to understand what values the variable can take, as well as the distribution and dispersion. This can be achieved with a histogram:

```python
sns.set(style='whitegrid', palette="deep", font_scale=1.1, rc={"figure.figsize": [8, 5]})
sns.distplot(data['x'], norm_hist=False, kde=False, bins=20, hist_kws={"alpha": 1}).set(xlabel='X coordinate', ylabel='Count');
```

With just one method sns.set(), we are able to style our figure, change the color, increase font size for readability, and change the figure size.
We can see that the most of the draw had being done for 4000 to 6000.

Create histograms for all of numerical variables with pandas offers:

```python
data[numerical].hist(bins=15, figsize=(15, 6), layout=(2, 3));
```

## Categorical variables
The values, distribution, and dispersion of categorical variables are best understood with bar plots.  Let’s analyze the _label_ variable which indicate which subjects has the disease and which of them are disease free subjects.

```python
sns.countplot(labels);
```

From the visualization, we can easily see that the dataset is balanced with 23 ET and 27 CT
Let's now to visualize all categorical variables

```python
fig, ax = plt.subplots(1, 2, figsize=(20, 10))
for variable, subplot in zip(categorical, ax.flatten()):
    sns.countplot(data[variable], ax=subplot)
    for label in subplot.get_xticklabels():
        label.set_rotation(90)
```

The most of the data is happen with pen down. Can be interesting see the relationship between both pen_up and label

## Analyzing relationships

### Numerical
Plotting relationships between variables allows us to easily get a visual understanding of patterns and correlations. The scatter plot is often used for visualizing relationships between two numerical variables. The seaborn method to create a scatter plot is very simple:

```python
sns.scatterplot(x=data['x'], y=data['y'], hue=data['labels']);
```

seaborn also provides us with a nice function called jointplot which will give you a scatter plot showing the relationship between two variables along with histograms of each variable in the margins — also known as a marginal plot.

```python
sns.jointplot(data=data, x='x', y='y');
```
