Configurar matplotlib para ver las gráficas en linea

```python
%matplotlib inline
import matplotlib.pyplot as plt
import warnings
import os
import h5py
import numpy as np
warnings.filterwarnings('ignore')
from nilearn.image.image import mean_img
from nilearn import (image, plotting)
```

Atlas del cerebelo

```python
fsl_prob_cerebelo = "/opt/fsl-5.0.10/data/atlases/Cerebellum/Cerebellum-MNIfnirt-prob-1mm.nii.gz"
```

Localización de los ficheros de neuroimagen de ET y Controles

```python
data_path = '/home/elekin/datos-niix/preproc/'
files = list(filter(lambda dir: '_session_id_1_subject_id' in dir,os.listdir(data_path)))
node = 'remove_noise'
file_name = 'detrend_regfilt.nii.gz'

fmri_files_path = np.array(list(map(lambda x: os.path.join(data_path,x,node,file_name),files)))
```

```python
fmri_files_path.shape
```

```python
from nilearn.image import resample_to_img

resampled_fmri_img = resample_to_img(fsl_prob_cerebelo,fmri_files_path[0])
```

```python
import nibabel as nb
fmri_img = nb.load(fmri_files_path[0])
resampled_fmri_img.shape,fmri_img.shape
```

```python
plotting.plot_prob_atlas(resampled_fmri_img, view_type='filled_contours')
plotting.plot_prob_atlas(fmri_img, view_type='filled_contours')
```

```python
# Compute the mean EPI: we do the mean along the axis 3, which is time
fmri_mean_img = mean_img(fmri_img)

from nilearn.plotting import plot_epi, show
display_epi = plot_epi(fmri_mean_img,cmap="gray",cut_coords=(2,-1,20))
display_epi.add_overlay(mean_img(resampled_fmri_img), cmap=plotting.cm.purple_green, threshold=3)
```

NiftiMapsMasker para extraer las señales del cerebelo. NiftiMapsMasker es útil cuando se deben extraer datos de volúmenes superpuestos 

```python
from nilearn.input_data import NiftiMapsMasker
masker = NiftiMapsMasker(maps_img=resampled_fmri_img,memory='nilearn_cache', verbose=True, standardize=True)
```

Overview de algunos indicadores de las imagenes

```python
n_regions = resampled_fmri_img.shape[-1]
n_regions
```

Entrenamiento del modelo para aplicar la extracción del cerebelo

```python
masker.fit()
```

```python
masker_extracted_img = masker.maps_img_
# Total number of regions extracted
masker_n_regions_extracted = masker_extracted_img.shape[-1]
  # Visualization of region extraction results
title = ('%d regions are extracted from cerebellum atlas'
           % (masker_n_regions_extracted)) 
plotting.plot_prob_atlas(masker.maps_img, view_type='filled_contours',
                           title=title
                           )
```

Extracción de las series temporales dado el atlas de probabilidad del cerebelo a partir de las imágenes fmri .

```python
h5_ts_file = h5py.File('elekin_phd_et.hdf5','w')
ts_group = h5_ts_file.create_group('cerebellum_ts','w')


masker_timeseries_each_subject = []
for i,filename in enumerate(fmri_files_path):
    ts=masker.transform(filename,confounds=None)
    masker_timeseries_each_subject.append(ts)
    ts_group.create_dataset(files[i],data=np.array(ts),compression="gzip", compression_opts=9)
h5_ts_file.close()
```

```python
masker_timeseries_each_subject = np.array(masker_timeseries_each_subject)[3]
masker_timeseries_each_subject.shape
```

Gráfica de las 28 series temporales extraidas

```python
for i in range(3):
    fig = plt.figure()
    plt.plot(masker_timeseries_each_subject[i])
    plt.xlabel('')
    plt.ylabel('')
    plt.show()
    plt.close()
```
